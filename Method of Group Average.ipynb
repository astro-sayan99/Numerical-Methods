## Method of Group Average

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.5, 6.7, 9.1, 10.8, 12.3, 14.8]

x1 = x[:4]
y1 = y[:4]
x2 = x[4:]
y2 = y[4:]

avg_x1 = sum(x1) / len(x1)
avg_x2 = sum(x2) / len(x2)
avg_y1 = sum(y1) / len(y1)
avg_y2 = sum(y2) / len(y2)

slope = (avg_y1 - avg_y2) / (avg_x1 - avg_x2)
intercept = (avg_y1 * avg_x2 - avg_y2 * avg_x1) / (avg_x2 - avg_x1)

y_fit = [slope * i + intercept for i in x]

plt.plot(x, y, 's--', color='red', lw=0.5, ms=5, label='Data Points')
plt.plot(x, y_fit, label=f"Fit: y = {slope:.4f}x + {intercept:.4f}")
plt.xlabel('x', fontdict={'family':'serif', 'color':'blue', 'fontsize':15})
plt.ylabel('y', fontdict={'family':'serif', 'color':'blue', 'fontsize':15})
plt.legend(shadow=True, prop={'family':'serif', 'size':12})
plt.grid()
plt.show()
```
