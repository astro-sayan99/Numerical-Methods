## Method of Moment

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([3.5, 3.5, 6.7, 9.1, 10.8, 12.3, 14.8])

h = (x[1] - x[0])
alpha = x[0] - (h / 2)
beta = x[-1] + (h / 2)

a11 = beta - alpha
a12 = (beta**2 - alpha**2) / 2
a21 = (beta**2 - alpha**2) / 2
a22 = (beta**3 - alpha**3) / 3
c11 = sum(y)
c22 = sum(i*j for i, j in zip(x, y))

D1 = np.array([[c11, a12], [c22, a22]])
D2 = np.array([[a11, c11], [a21, c22]])
D3 = np.array([[a11, a12], [a21, a22]])

a = np.linalg.det(D1) / np.linalg.det(D3)
b = np.linalg.det(D2) / np.linalg.det(D3)

y_fit = a + b*x

plt.plot(x, y, 's--', color='red', lw=0.5, ms=5, label='Data Points')
plt.plot(x, y_fit, label=f"Fit: y = {a:.4f} + {b:.4f}x")
plt.xlabel('x', fontdict={'family':'serif', 'color':'blue', 'fontsize':15})
plt.ylabel('y', fontdict={'family':'serif', 'color':'blue', 'fontsize':15})
plt.legend(shadow=True, prop={'family':'serif', 'size':12})
plt.grid()
plt.show()
```
